```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
```

```python
new_df = df.drop('per_capita_income',axis='columns')
new_df
```

```python
per_capita_income = df.per_capita_income
per_capita_income
```

### Create linear regression object

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2021]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = m * x + b
2020*828.46507522 + -1632210.7578554575
```

```python
a = reg.predict(new_df)
a
```

```python
new_df['per_capita_income'] = a
new_df
```

```python
new_df.to_csv("canada_per_capita_income_updated")
```
